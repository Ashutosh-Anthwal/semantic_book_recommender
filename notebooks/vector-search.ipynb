```python
from langchain_community.document_loaders import TextLoader  # langchain to connect different ai systems into working system.
from langchain_text_splitters import CharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma
```

```python
import pandas as pd
books = pd.read_csv("books_cleaned.csv")
```

```python
books
```

```python
books["tagged_description"].to_csv("tagged_description.txt",
                                   sep = "\n",
                                   index = False,
                                   header = False)
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter

raw_documents = TextLoader(
    "../tagged_description.txt",
    encoding="utf-8"
).load()

text_splitter = CharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50,
    separator="\n"
)

documents = text_splitter.split_documents(raw_documents)
```

```python
print(len(documents))
```

```python
print(documents[0])
```

```python
embedding_model = HuggingFaceEmbeddings(
    model_name="all-MiniLM-L6-v2"
)

db_books = Chroma.from_documents(
    documents,
    embedding=embedding_model
)
```

```python
query = "A book to teach children about nature"

docs = db_books.similarity_search(query, k=10)

docs
```

```python
books[books["isbn13"] == int(docs[0].page_content.split()[0].strip())]
```

```python
def retrieve_semantic_recommendations(
        query: str,
        top_k: int = 10,
) -> pd.DataFrame:

    recs = db_books.similarity_search(query, k=50)

    books_list = []

    for rec in recs:
        isbn = int(rec.page_content.strip('"').split()[0])
        books_list.append(isbn)

    return books[books["isbn13"].isin(books_list)]
```

```python
retrieve_semantic_recommendations("a book about world war")
```
